### <a name='1'></a> Wykres liniowy


````
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
````

````
fig = plt.figure()
fig.show()
````

````
_ = plt.plot([0, 1], [0, 3])
````

````
plt.plot([0, 1], [0, 3], label = 'line', color = 'green')
plt.xlabel('os x')
plt.ylabel('os y')
plt.title('wykres')
plt.legend()
````

````
plt.plot([0, 1, 2], [0, 3, 2], label = 'line', color = 'green')
plt.xlabel('os x')
plt.ylabel('os y')
plt.title('wykres')
plt.legend()
````

````
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
plt.plot(x, x_cos)
````

### <a name='2'></a> Tworzenie wielu wykresow

````
plt.subplot(121)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(122)
plt.plot([1, 1, 2], [3, 3, 5], 'r--')
````

````
fig = plt.figure(figsize = (12, 9))

plt.subplot(131)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(132)
plt.plot([1, 1, 2], [3, 3, 5], 'r--')
plt.subplot(133)
plt.plot([1, 1, 2], [1, 3, 5], 'ro')

fig.show() # trzeba tego uzywac w programach, w notebookach automatycznie wypluwa
````
